```python
import pandas as pd
import json
import ast
import re
import numpy as np
```

```python
#  convirtiendo la lista de objetos JSON en un diccionario en Python y luego a  DataFrame :

with open('steam_games.json/output_steam_games.json', 'r', encoding='utf-8') as archivo:
    data = [json.loads(line) for line in archivo]
    
df_outputsteam_games = pd.DataFrame(data)
```

```python
#Eliminando filas con valores faltantes :

df_outputsteam_games.dropna(how='all', inplace=True)
```

```python
df_outputsteam_games.info()
```

```python
df_outputsteam_games.head(5)
```

```python
# normalizando 

def normalize_developer(name):
    if isinstance(name, str):  
        cleaned_name = re.sub(r'[^a-zA-Z\s]', '', name)
        normalized_name = re.sub(r'\s+', ' ', cleaned_name.lower().strip())
        return normalized_name
    else:
        return name

df_outputsteam_games['publisher'] = df_outputsteam_games['publisher'].apply(normalize_developer)
```

```python
df_outputsteam_games['developer'] = df_outputsteam_games['developer'].apply(normalize_developer)
```

```python
df_outputsteam_games.tail(5)
```

```python
#revisando valores faltantes:

df_outputsteam_games.isna().sum()
```

```python
# llenando valores faltantes en la columna title:

df_outputsteam_games['title'].fillna(df_outputsteam_games['app_name'], inplace=True)
```

```python
# llenando valores faltantes en la columna genres:

df_outputsteam_games['genres'].fillna(df_outputsteam_games['tags'], inplace=True)
```

```python
#revisando valores en columna precio: 
df_outputsteam_games['price'].unique()
```

```python
df_outputsteam_games.isna().sum()
```

```python
#eliminando valores faltantes de la columna price: 

df_outputsteam_games.dropna(subset='price', inplace=True)

#comprobando valores faltantes en columna price:

df_outputsteam_games['price'].isnull().sum()
```

```python
#reemplazando cadena de caracteres que contengan Free a valor 0.00:

df_outputsteam_games['price'] = df_outputsteam_games['price'].replace(r'.*Free.*', '0.00', regex=True)
df_outputsteam_games
```

```python
# conviertiendo la columna price a formato numerico y  los que contengan NaN a 0.00

df_outputsteam_games['price'] = pd.to_numeric(df_outputsteam_games['price'], errors='coerce')
df_outputsteam_games['price'].fillna('0.00', inplace=True)
```

```python
df_outputsteam_games['price'].unique()
```

```python
#convirtiendo cadena de caracteres a numericos:

df_outputsteam_games['price'] = pd.to_numeric(df_outputsteam_games['price'], errors='coerce').astype('float32')

```

```python
#convirtiendo numeros con decimales a 2 decimales:

np.set_printoptions(formatter={'float_kind':'{:0.2f}'.format})
np.round(df_outputsteam_games['price'].unique(), 2)
```

```python
df_outputsteam_games['price'].unique()
```

```python
# extrayendo el anio de cada fecha en la columna release_date:

df_outputsteam_games['release_date'] = df_outputsteam_games['release_date'].str.extract(r'(\d{4})')
```

```python
df_outputsteam_games['release_date'].unique()
```

```python
#eliminando los valores NaN de la columna release_date:

df_outputsteam_games.dropna(subset=['release_date'], inplace=True)
```

```python
df_outputsteam_games['release_date'].unique()
```

```python
# eliminando columnas que no son relevantes:

df_outputsteam_games.drop([ 'url', 'tags', 'reviews_url', 'specs', 'early_access'], axis=1, inplace=True)
```

```python
df_outputsteam_games.drop([ 'title'], axis=1, inplace=True)
```

```python
# eliminando valores NaN
df_outputsteam_games.dropna(subset=['id'], inplace=True)
```

```python
# cambiando tipo de datos:

df_outputsteam_games['release_date'] = df_outputsteam_games['release_date'].astype('int')
df_outputsteam_games['id'] = df_outputsteam_games['id'].astype('int')
df_outputsteam_games['developer'] = df_outputsteam_games['developer'].astype('string')
df_outputsteam_games['app_name'] = df_outputsteam_games['app_name'].astype('string')
df_outputsteam_games['publisher'] = df_outputsteam_games['publisher'].astype('string')
```
